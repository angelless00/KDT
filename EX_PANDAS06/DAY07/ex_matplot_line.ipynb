### [ Matplotlib을 활용한 시각화 ]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
# -  패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기
plt.plot([1,2,3,4])
plt.xlabel('Data')          # 반환값  Text(0.5, 0, 'Data')
plt.ylabel('Result')        # 반환값 Text(0, 0.5, 'Result')
plt.show() ## 1대1 대화방식이라 반환값들이 존재! 반환값들 안보이고 최종값들만 보이게!
```

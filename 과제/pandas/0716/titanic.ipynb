```python
import pandas as pd
import numpy as np
titanicDF=pd.read_csv('titanic.csv')
```

```python
titanicDF.info()
```

```python
# 결측치확인
titanicDF.isna().sum()
```

```python
# 중복치 확인
titanicDF.duplicated().sum()
```

```python
# 중복 데이터 삭제
titanicDF.drop_duplicates()
```

```python
# 공란이 너무 많은 deck 컬럼삭제
titanicDF.drop('deck',axis=1,inplace=True)

# 의미가 같은 surived 와 alive 에서 alive 삭제
titanicDF.drop('alive',axis=1,inplace=True)

# 의미가 같은 pclass와 class에서 class 삭제
titanicDF.drop('class',axis=1,inplace=True)

# 의미가 같은 embarked 와 embark_town에서 enbarked삭제
titanicDF.drop('embarked',axis=1,inplace=True)

# 의미가 겹치는 sex와 who중 sex 삭제
titanicDF.drop('sex',axis=1,inplace=True)

# 의미가 겹치는 who 와 adlut_male중 adult_male삭제
titanicDF.drop('adult_male',axis=1,inplace=True)
```

```python
# age데이터 nan값 처리
manave=titanicDF[titanicDF['who']=='man']['age'].mean()
womanave=titanicDF[titanicDF['who']=='woman']['age'].mean()
chilave=titanicDF[titanicDF['who']=='child']['age'].mean()
for i in titanicDF.index:
    if pd.isna(titanicDF.loc[i,'age']):
        if titanicDF.loc[i,'who']=='man':
            titanicDF.loc[i,'age']=manave
        elif titanicDF.loc[i,'who']=='woman':
            titanicDF.loc[i,'age']=womanave
        else:
            titanicDF.loc[i,'age']=chilave
```

```python
# 결측치 확인
titanicDF.isna().sum() 
```

```python
# 전체데이터에 영향없는 embark_town 결측치 삭제
titanicDF.dropna(inplace=True)
titanicDF.isna().sum()
```

```python
# 나이에 따라 어린이 청년 중년 노인으로 분류
age_range=pd.cut(titanicDF['age'],bins=4,labels=['어린이','청년','중년','노인'])
titanicDF['age_range']=age_range
```

```python
# 사망자데이터를 tatan_Death에
# 생존자데이터를 tntan_Survive에
titan_Death=titanicDF[titanicDF['survived']==0]
titan_Survive=titanicDF[titanicDF['survived']==1]
```

```python
# 타이타닉 탑승자 전체 데이터
titanicDF.describe(include='all')
```

```python
# 타이타닉 생존자 전체 데이터
titan_Death.describe(include='all')
```

```python
# 타이타닉 사망자 전체 데이터
titan_Survive.describe(include='all')
```

```python
# 데이터 카테고리화
#titanicDF['pclass']=titanicDF['pclass'].astype('category')
titanicDF['who']=titanicDF['who'].astype('category')
titanicDF['embark_town']=titanicDF['embark_town'].astype('category')
```

```python
# 'pclass','who','embark_town','age_range,alone'를 [전체,생존자,사망자]순으로 카운트
pclass=pd.concat([titanicDF['pclass'].value_counts(),titan_Survive['pclass'].value_counts(),titan_Death['pclass'].value_counts()],axis=1,keys=['Total','Survive','Death'])
who=pd.concat([titanicDF['who'].value_counts(),titan_Survive['who'].value_counts(),titan_Death['who'].value_counts()],axis=1,keys=['Total','Survive','Death'])
embark_town=pd.concat([titanicDF['embark_town'].value_counts(),titan_Survive['embark_town'].value_counts(),titan_Death['embark_town'].value_counts()],axis=1,keys=['Total','Survive','Death'])
age_range=pd.concat([titanicDF['age_range'].value_counts(),titan_Survive['age_range'].value_counts(),titan_Death['age_range'].value_counts()],axis=1,keys=['Total','Survive','Death'])
alone=pd.concat([titanicDF['alone'].value_counts(),titan_Survive['alone'].value_counts(),titan_Death['alone'].value_counts()],axis=1,keys=['Total','Survive','Death'])
```

```python
pclass['Survive_ratio']=pclass['Survive']/pclass['Total']
pclass.sort_index()
```

```python
embark_town['Survive_ratio']=embark_town['Survive']/embark_town['Total']
embark_town
```

```python
age_range['Survive_ratio']=age_range['Survive']/age_range['Total']
age_range
```

```python
#p.31 데이터의 분위수를 구하는 quantile 메서드
df=pd.read_csv(r'C:\KDT\과제\시각화\datasets\bookings\Bookings.csv')
df['Total_Review']=df['Total_Review'].map(lambda x:str(x).replace('external','').strip())
df['Total_Review']=df['Total_Review'].map(lambda x:str(x).replace('review','').strip())
df['Total_Review']=df['Total_Review'].map(lambda x:str(x).replace(',',''))
df['Total_Review']=df['Total_Review'].astype('float')
```

```python
quantille=[0,0.2,0.4,0.6,0.8,1]
for idx in quantille:
    q=df['Total_Review'].quantile(idx,interpolation='lower')
    print(f'quantille({idx}) is {q}')
```

```python
for idx in quantille:
    q=df['Total_Review'].quantile(idx,interpolation='nearest')
    print(f'quantille({idx}) is {q}')
```

```python
#p.37 pandas의 꽃, 그룹별 연산을 위한 groupby 메서드
df=sns.load_dataset('titanic')
df.head()
```

```python
df.groupby('sex')['survived'].mean()
```

```python
#df.groupby(['sex','class'])['survived'].mean()#
df.groupby(['sex','class'],observed=False)['survived'].mean()
```

```python
df.groupby(['sex','class'],observed=False)['survived'].agg(['mean','count'])
```

```python
df.groupby(['sex','class'],observed=False)[['survived','age']].agg({'survived':'mean','age':'max'})
```

```python
def get_IQR(data):
    _3rd=data.quantile(0.75)
    _1st=data.quantile(0.25)
    return (np.abs(_3rd-_1st)*1.5)

df.groupby(['sex','class'],observed=False)['age'].apply(get_IQR)
```

```python
import pandas as pd 
df1=pd.read_csv('test_score.csv',encoding='cp949')
df1.dropna(inplace=True)
max_age=df1['연령대'].value_counts().idxmax()
score=df1[(df1['연령대']==max_age)&(df1['과목명']=='기본간호학')]['과목별점수']
round((score.max()-score.mean())/score.std(),4)
```

```python
df2=pd.read_csv('diabetes.csv',encoding='cp949')
col=df2.corr()['progression'].sort_values(ascending=False).index[1]
round(df2[col].max(),2)
```

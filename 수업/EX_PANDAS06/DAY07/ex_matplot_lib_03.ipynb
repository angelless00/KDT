### [ Matplotlib 활용 데이터 시각화 ]
- X축과 Y축 다양한 처리

```python
# [1] 모듈 로딩
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt  같은놈이야!!!
```

```python
# [2] 그래프 그리기 - 기본
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,'o-')
plt.show()
```

```python
# [2] 그래프 그리기 - x축과 y축 범위 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,'o-')
plt.ylim(bottom=0,top=15)
#plt.xlim(left='2024-01-01',right=2024-05-01')  ## 문자 못받아!
plt.xlim(left=-1,right=10)
plt.xticks(rotation=45)   # x축 스케일 기울여줘!
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 출력 창 즉, figure 크기 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.figure(figsize=(20,5))          # 그래프 사이즈 키워줘
plt.plot(x_data,y_data,'o-')
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 출력 창 즉, figure를 자동맞춤
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,'o-')
plt.tight_layout()      # 모든 내용이 출력 될 수 있도록 자동 맞춤 설정
plt.show()
```

```python
# [3] 그래프 그리기 - 눈금(ticks) 설정
y_data=[1,3,5,7,9]


plt.plot(y_data,'o-')
plt.xlim(left=-1, right=7)
plt.ylim(bottom=0,top=15)
plt.xticks([2,4,6,8],['A','B','C','D'],fontsize=22,fontweight='bold')
# 이렇게도 설정 가능해!
# font_config={'fontsize':22,'fontweight':'bold'}
# plt.xticks([2,4,6,8],['A','B','C','D'],**font_config)

plt.show()
```

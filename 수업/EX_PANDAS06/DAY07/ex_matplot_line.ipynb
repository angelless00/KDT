### [ Matplotlib을 활용한 시각화 ]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
# -  패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(y축데이터)
plt.plot([1,2,3,4])
plt.xlabel('Data')          # 반환값  Text(0.5, 0, 'Data')
plt.ylabel('Result')        # 반환값 Text(0, 0.5, 'Result')
plt.show() ## 1대1 대화방식이라 반환값들이 존재! 반환값들 안보이고 최종값들만 보이게!
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(x축데이터,y축데이터)
plt.plot([10,20,30],[5,10,20])
plt.show() 
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(x축데이터,y축데이터, fmf(포맷스트링))
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,20],'r--')
plt.show() 
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(x축데이터,y축데이터, fmf(포맷스트링))
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
# 색상코드 r g b k
# 점모양코드 o ^ > < +
# 선모양코드 - --
#plt.plot([10,20,30],[5,10,20],'ro--') # 동그라미
#plt.plot([10,20,30],[5,10,20],'b^--') # 위로가는 삼각형
plt.plot([10,20,30],[5,10,20],'k-+') # 오른쪽방향 삼각형
plt.show() 
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(x축데이터,y축데이터,line2D속성으로 설정)
# fmf 포맷스트링 => Line2D의 속성으로 지정
# color='red', marker='o', linestyle='dashed',markersize=10,linewidth=5
# mec=marker edge color
# 다 못외우니 matplotlib 홈페이지 참조
plt.plot([10,20,30],[5,10,20],color='c', marker='o', linestyle='dashed',markersize=10,linewidth=5,mec='blue',mew=5)
plt.show()
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plt.plot([10,20,30],[15,25,35],'bo')
# plt.plot([10,20,30],[18,29,25],'ro')
plt.plot([10,20,30],[15,25,35],'bo-',[10,20,30],[18,29,25],'ro-') # 한번에 적기도 가능
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'bo--',label='V')
plt.plot([10,20,30],[15,30,25],'ro--',label='2')
plt.title('[MPLIB LINE2D]')
plt.xlabel('Data')
plt.ylabel('Result')
#plt.legend('values')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함
#                            loc속성으로 위치 지정 가능 [기본값:best]
plt.plot([10,20,30],[15,25,20],'bo--',label='V')
plt.plot([10,20,30],[15,30,25],'ro--',label='2')
plt.title('[MPLIB LINE2D]')
plt.xlabel('Data')
plt.ylabel('Result')
#plt.legend('values')
plt.legend(loc='center right')
# lower right , upper center 등등이 있음
plt.show()
```

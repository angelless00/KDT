```python
# [1-1] 모듈로딩
import pandas as pd                     # 데이터 분석용 모듈
import matplotlib.pyplot as plt         # 데이터 시각화용 모듈
import numpy as np
```

```python
# [1-2] 데이터 준비
F_data=r'C:\KDT\data\시도별 전출입 인구수.xlsx'
```

```python
# [1-3] DATA => DataFrame 형태 저장
dataDF=pd.read_excel(F_data,engine='openpyxl')
```

- [2] 데이터 확인 : 행과 컬럼수, 실제데이터와 컬럼 타입, 결측치 체크

```python
# [2-2] 기본 정보 확인 -> info(),describe()
# - 수치 데이터 컬럼의 데이터 분포 확인 : 최대값, 최소값, 표준편차, 사분위수, 평균값
# - 범주 데이터 컬럼의 데이터 분포 확인 : 고유값, 최빈값, 최빈값의 빈도수
dataDF.describe(include='all')
```

```python
### 기본 정보 확인 후 결과
# - 불필요한 행/컬럼 : 0번행 삭제
# - 결측치 : '전출지별' 컬럼 => 이전 행값으로 채우기
# - 이상치 : 행별로 '-' 값이 수치값과 섞여있음 => 0 처리 필요!
```

- [3] 데이터 전처리
    - 기본확인 파악된 내용들 처리
    - 불필요 데이터 삭제 : 0번행 삭제
    - 결측치 데이터 처리 : '전출지별' 컬럼 처리
    - 이상치 데이터 처리 : '-'에 대한 처리
    - 데이터 자료형 일치 : 형변환


```python
# [3-1] 불필요 데이터 처리
copyDF=dataDF.drop(index=0)
print(f'확인\n {copyDF.head(3)}')
```

```python
# [3-2] 결측치 데이터 처리 => isna()/isnull()
copyDF.isna().sum()
```

```python
# - '1984' '1985' '1986' 컬럼에서 결측치 데이터만 추출
a=copyDF[[1984,1985,1986]].isna()
a[a.sum(axis=1)>0]
```

```python
# - '전출지별' 
copyDF['전출지별']=copyDF['전출지별'].ffill()

# - 수치데이터 결측치 처리
pd.set_option('future.no_silent_downcasting', True)
copyDF=copyDF.fillna('0').copy()        # 경고처리 copyDF
```

```python
# [3-3] 이상치 데이터 처리
copyDF.replace('-',0,inplace=True)
```

```python
# [3-4] 자료형 변환 : 48개의 컬럼을 int로 형변환
d_=dict(zip(copyDF.columns[2:].to_list(),['int']*(copyDF.shape[1]-2)))
#copyDF=copyDF.astype({1970:'int',1971:'int'........})<< 하기힘들어서!
copyDF=copyDF.astype(d_)
```

- [4] 데이터 가공  
    - 대구에서 전지역 전입 인구수 분석
    - 전출지별 - '대구광역시'
    - 전입지별 - '대구광역시', '전국' 제외

```python
# 대구광역시 데이터만 추출
mask = (copyDF['전출지별']=='대구광역시')
cityDF=copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF=cityDF.drop('전출지별',axis=1)

# - 전입지별 컬럼행 인덱스 설정
cityDF.set_index('전입지별',inplace=True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['대구광역시','전국'],inplace=True)
```

```python
# - 가공데이터 확인
print(cityDF.dtypes,cityDF.columns,cityDF.index,sep='\n\n')
```

- [5] 히스토그램으로 특정 컬럼 데이터 분포 시각화

```python
# - 2000년도 데이터의 분포
# 값 기준 정렬 ==> [기본] 오름차순
cityDF[2000].sort_values()
```

```python
# 2000년도 대구지역에서 타지역으로 이동 인구수 분포
plt.hist(cityDF[2000],bins=20,histtype='stepfilled')    # bins의 기본값 10
plt.hist(cityDF[1990],bins=20,histtype='step')    
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
ax1,ax2=fig.subplots(1,2,sharex=True)
ax1.hist(cityDF[2000],bins=20,histtype='stepfilled') 
ax2.hist(cityDF[1990],bins=20,histtype='step')    
plt.show()
```

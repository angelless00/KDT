### [ 파이 그래프 ]
- 컬럼(변수)의 분표 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비 
file=r'C:\KDT\EX_PANDAS06\DAY04\auto_mpg.csv'
```

```python
# [3] CSV => DataFrame으로 저장
dataDF=pd.read_csv(file)
```

```python
# - 컬럼별 타입 확인
dataDF.info()
```

```python
# 한글 폰트 설정 => 폰트 매니저 모듈 사용
from matplotlib import font_manager as fm
from matplotlib import rc

# 적용할 폰트 파일
Font_file=r'C:\Windows\Fonts\gulim.ttc'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=Font_file).get_name()

# 새로운 폰트 패밀리 이름지정
rc('font',family=font_name)
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산 국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar그래프
plt.bar(dataDF['origin'].unique(),dataDF['origin'].value_counts())
```

```python
## 강사님 코드
originSR=dataDF['origin'].value_counts()
x=originSR.index
y=originSR.values
```

```python
originbar=plt.bar(x,y)
plt.xticks([1,2,3],['미국','유럽','일본']) # 1,2,3 눈금 위치값 
plt.bar_label(originbar,y,label_type='center')
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(y,labels=['미국','유럽','일본'],autopct='%.2f%%',startangle=90,explode=[0,0,0.1])
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(y,labels=['미국','유럽','일본'],autopct='%.2f%%',wedgeprops={'linewidth':1,'width':0.5,'edgecolor':'w'})

plt.show()
plt.show()
```

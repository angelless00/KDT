=> 지도/비지도/강화 => 학습알고리즘
### 학습알고리즘 
- 선형모델 
    - 피쳐와 타겟의 관계가 선형관계
    - 정규분포 기반
    - 선형식 : $ y=ax+b $=> $wx+b$ => $ w_0+w_1x $  
            : $ y=w_0 +w_1x +w_2x+....+w_nx $
    - Coefficient 또는 Weight : 학습을 통해서 다양한 피쳐들의 영향정도를 정확하게 반영하기 위해 사용됨
    - 피쳐가 모두 적용되게하면 과대적합이 될 확률 high => 규제가 필요! 가중치를 조절하여 규제
        - Lasso : 1-norm -> 가중치가 0으로 갈 확률 high
        - Ridge : 2-norm
        - ElasticNet : Lasso와 Ridge 복합
        
### 파라미터
- 모델 파라미터 
    - fit() 학습 진행 후 설정되는 파라미터
    - 모델.파라미터명_
- 하이퍼 파라미터
    - 모델 인스턴스 생성시 매개변수
    - 모델 인스턴스 생성할때 개발자가 값을 설정
    - 모델 성능에 영향을 미침
    - ex> 규제강도 alpha, k 개수, 학습 횟수 max_iter, 규제강도 C (alpha의 역수)
- 튜닝 : 모델 성능을 높이는 작업 즉, 모델의 하이퍼파라미터의 값을 조절해서 성능 올리기

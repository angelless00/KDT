국내 전통주를 보기위해 민속주와 특산주를 합치겠음

```python

pd.concat([major_alcho5['증류식소주'],oversea_alcho5['소주'],juntong_alcho5['증류식소주']],axis=1,keys=['Total','oversea','juntong'])
```

해외의경우 증류식 소주의 데이터가 없어서 소주데이터로 대체.
국내 전통주시장의 증류식 소주출고가 늘어난것으로 보임
전통주의 소비 증가 => 술담화 같은 전통주 구독서비스의 탄생
전체 전통주는 어떻게 변화하였는가

```python
## 전통주 출고량 변화 2018-2020
juntong_change=juntong_alcho5.loc[2020]-juntong_alcho5.loc[2018]
juntong_change_ratio=juntong_alcho5.loc[2020]/juntong_alcho5.loc[2018]*100
pd.concat([juntong_change,juntong_change_ratio],axis=1,keys=['전통주소비변화량(kl)','전통주소비변화율(%)'])
```

전통주는 수입에 많이 의존하는 과실주를 제외하곤 전체적으로 소비량이 강하게 증가한것으로 볼수있다

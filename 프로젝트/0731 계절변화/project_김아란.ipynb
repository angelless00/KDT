```python
def before(DF):
    DF.replace({' ':pd.NA,'관측 안됨':pd.NA,'결측':pd.NA},inplace=True)
    print(DF.isna().sum(axis=0))
```

```python
before(cosmosDF)
# 결측치가 적은 개화시기를 기준으로 판단
```

```python
# 결측컬럼 제거
cosmosDF.drop(['발아','발아(평비)','만발','만발(평비)'],axis=1,inplace=True)
```

```python
# 결측치 제거후 날짜 datetime 변경, valuecount
def before1(DF,column):
    DF.dropna(inplace=True)
    DF[column]=pd.to_datetime(DF[column])
    print(DF[('지점', 'Unnamed: 0_level_1')].value_counts())
```

```python
def preper_plot(DF,korean_name,column):
    cosmos_region=DF[DF[('지점', 'Unnamed: 0_level_1')]==korean_name]
    cosmos_region.loc[:,column]=pd.to_datetime({'year':2020,'month':cosmos_region[column].dt.month,'day':cosmos_region[column].dt.day})
    return cosmos_region
```

```python
def plot_sub(DF,i,korean_name,column,colors):
    ax[i].plot(preper_plot(DF,korean_name,column)[('년도', 'Unnamed: 1_level_1')],preper_plot(DF,korean_name,column)[column],color=colors)
    ax[i].plot([2004,2023],[preper_plot(DF,korean_name,column)[column].min(),preper_plot(DF,korean_name,column)[column].max()],'k--')
    ax[i].set_title(korean_name,fontsize=10)
    ax[i].set(xticks=list(range(2004,2024,4)))
    ax[i].grid(True,linewidth=0.5)
```

```python
before1(cosmosDF,'개화')
```

```python
cosmosregion=['백령도','북강릉','대전','여수','포항']
color=['r','b','g','orange','purple']
```

```python
fig=plt.figure(figsize=(25,5))
ax=fig.subplots(1,5,sharey=True)
fig.suptitle('지역별 코스모스 개화시기',fontsize=20)
for i in range(5):
    plot_sub(cosmosDF,i,cosmosregion[i],'개화',color[i])
plt.show()
```

```python
# 인천앞바다 수온변화
```

```python
seatemp1=pd.read_csv('OBS_BUOY_DD_2005~14.csv',encoding='euc-kr')
seatemp2=pd.read_csv('OBS_BUOY_DD_2015~23.csv',encoding='euc-kr')
```

```python
seatemp1.info()
```

```python
seatemp1['일시']=pd.to_datetime(seatemp1['일시'])
seatemp2['일시']=pd.to_datetime(seatemp2['일시'])
```

```python
seatemp1=seatemp1[(seatemp1['일시'].dt.month==10 )|(seatemp1['일시'].dt.month==11)]
seatemp2=seatemp2[(seatemp2['일시'].dt.month==10 )|(seatemp1['일시'].dt.month==11)]
```

```python
seatemp=pd.concat([seatemp1,seatemp2],axis=0)
```

```python
seatemp.info()
```

```python
seatemp.head()
```

```python
seatemp.tail()
```

```python
tempmean=[]
for i in range(2005,2024):
    tempmean.append(seatemp[seatemp['일시'].dt.year==i]['평균 수온(°C)'].mean())
```

```python
plt.plot(list(range(2005,2024)),tempmean)
plt.title('연도별 10월,11월 인천앞바다 평균수온변화 그래프',fontsize=15)
plt.xlabel('연도')
plt.ylabel('평균수온')
```

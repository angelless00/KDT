```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

import collections
import RNNutils as utils

if not hasattr(collections,'callable'):
    collections.Callable=collections.abc.Callable


def music_chart(weeknum):
        """10월 첫째주부터 주간차트 과거로"""
        # 중복제거를 위해 set 형태로
        songid=set()
        for a in  range(weeknum):
            url=f'https://music.bugs.co.kr/chart/track/week/total?chartdate={pd.Timestamp(20000-7*a, unit="D").strftime("%Y%m%d")}'
            html=requests.get(url)
            soup=BeautifulSoup(html.text,'html.parser')
            # 링크 리스트를 뽑아냄
            songs=soup.find('tbody').find_all('tr')
            for song in songs:
                song_id=song.get('trackid')
                # set에 추가
                songid.add(song_id)
        # list 형태로 반환
        return list(songid)

import re
def get_ly(songid):
    # 가사
    lys=[]
    for id in songid:
        url=f'https://music.bugs.co.kr/track/{id}'
        html=requests.get(url)
        soup=BeautifulSoup(html.text,'html.parser')
        ly=soup.find('xmp')
        ly=str(ly).replace('<xmp>','').replace('</xmp>',"")
        ly=re.sub('[^가-힣a-zA-Z]+',' ',ly)
        lys.append(ly)
    return lys
```

```python
# 음악차트 속 코드 긁어오기
songid=music_chart(250)
len(songid)
```

```python
# 코드에 해당하는 가사 전처리후 긁어오기
get_ly(songid)
```
